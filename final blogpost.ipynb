# Graph average difference between fuel types:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://corgis-edu.github.io/corgis/datasets/csv/cars/cars.csv"
df = pd.read_csv(url)

# Extract relevant columns
fuel_data = df[['Fuel Information.Highway mpg', 'Fuel Information.Fuel Type']]

# Group data by Fuel Type and calculate the mean Highway mpg for each group
mean_highway_mpg = fuel_data.groupby('Fuel Information.Fuel Type')['Fuel Information.Highway mpg'].mean().reset_index()

# Sort data by mean Highway mpg in descending order
mean_highway_mpg = mean_highway_mpg.sort_values(by='Fuel Information.Highway mpg', ascending=False)

# Plot the comparison using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Fuel Information.Fuel Type', y='Fuel Information.Highway mpg', data=mean_highway_mpg, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Fuel Type')
plt.ylabel('Mean Highway mpg')
plt.title('Comparison of Fuel Type based on Mean Highway mpg')
plt.tight_layout()
plt.show()
```

# Graph elaborating previous graph, showing mean highway mpg for cars of each year separated by fuel type AND also transmission

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://corgis-edu.github.io/corgis/datasets/csv/cars/cars.csv"
df = pd.read_csv(url)

# Calculate the mean Highway mpg for cars with the same 'Identification.Year', 'Fuel Information.Fuel Type', and 'Identification.Classification'
mean_highway_mpg_by_year_fuel_auto_manual = df.groupby(['Identification.Year', 'Fuel Information.Fuel Type', 'Identification.Classification'])['Fuel Information.Highway mpg'].mean().reset_index()

# Plot the comparison using a line plot for each fuel type, split by automatic and manual transmissions
plt.figure(figsize=(12, 6))
sns.lineplot(data=mean_highway_mpg_by_year_fuel_auto_manual, x='Identification.Year', y='Fuel Information.Highway mpg', hue='Fuel Information.Fuel Type', style='Identification.Classification', markers=True, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Model Year')
plt.ylabel('Mean Highway mpg')
plt.title('Mean Highway mpg for Cars of Each Year, Fuel Type, and Transmission')
plt.legend(title='Transmission', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

Based on the analysis of the three graphs, it is evident that diesel fuel consistently achieves the highest miles per gallon (MPG) among the various fuel types. Specifically, diesel manual vehicles stand out as the most efficient, achieving approximately 30 miles per gallon or higher on average. In a close second place, compressed natural gas (CNG) exhibits competitive fuel efficiency, but it still falls behind diesel manual vehicles in terms of MPG performance.

It should be noted that CNG is exclusively available in automatic transmission, which, on average, results in higher MPG compared to the combined average of diesel manual and diesel automatic vehicles. This indicates that automatic CNG vehicles have relatively better fuel efficiency than their diesel automatic counterparts, though they still do not match the performance of diesel manual vehicles.

In contrast, E85 fuel ranks below all other fuel types in terms of MPG performance. Gasoline-powered vehicles occupy a middle ground, with their MPG ratings falling between those of automatic diesel and manual diesel vehicles. Specifically, gasoline manual vehicles tend to perform better in terms of MPG compared to automatic gasoline vehicles.

In conclusion, diesel manual vehicles consistently demonstrate the highest fuel efficiency, followed closely by automatic CNG vehicles. Gasoline-powered vehicles, while offering a decent MPG, tend to perform better in manual transmission and slightly worse in automatic transmission compared to their diesel counterparts. E85 fuel lags behind all other fuels in terms of MPG performance.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://corgis-edu.github.io/corgis/datasets/csv/cars/cars.csv"
df = pd.read_csv(url)

# Filter out rows where 'Engine Information.Engine Type' contains the word 'hybrid'
filtered_data = df[~df['Engine Information.Engine Type'].str.contains('hybrid', case=False)]

# Sort the filtered data by 'Fuel Information.Highway mpg' in descending order
sorted_data = filtered_data.sort_values(by='Fuel Information.Highway mpg', ascending=False)

# Select the top 30 cars based on 'Fuel Information.Highway mpg'
top_thirty_cars = sorted_data.head(30)

# Create the bar chart with different colors based on 'Fuel Information.Fuel Type' and 'Engine Information.Engine Type'
colors = []
for fuel_type, engine_type in zip(top_thirty_cars['Fuel Information.Fuel Type'], top_thirty_cars['Engine Information.Engine Type']):
    if 'diesel' in engine_type.lower() or fuel_type == 'Diesel':
        colors.append('lightgrey')
    elif fuel_type == 'Compressed Natural Gas':
        colors.append('green')
    else:
        colors.append('blue')

plt.figure(figsize=(14, 6))
plt.bar(top_thirty_cars['Identification.ID'], top_thirty_cars['Fuel Information.Highway mpg'], color=colors)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Car ID')
plt.ylabel('Highway mpg')
plt.title('Top 30 Cars with Highest Highway mpg (Color-Coded by Fuel and Engine Type)')
plt.tight_layout()
plt.show()
```

A comparative analysis of the top three vehicles exhibiting the best miles per gallon (MPG) reveals that 12 out of the 30 top-performing cars belong to the diesel category, with a specific focus on Volkswagen models. However, there are notable exceptions within this group, comprising the 2011 and 2012 Chevrolet Cruze eco, along with the 2012 Honda Civic HF sedan. Despite not falling under the diesel category, these vehicles still demonstrate exceptional fuel efficiency and secure commendable positions within the top 30 MPG rankings. Remarkably, all these cars achieve an MPG rating hovering around or slightly below the 40 mile per gallon mark.

It is important to emphasize that the analysis did not account for any hybrid vehicles or turbocharged cars in the top 30 MPG list. Furthermore, none of the vehicles in this ranking utilize compressed natural gas (CNG) or E85 fuel types.

In conclusion, while diesel-powered Volkswagen cars predominantly occupy the top three spots in terms of MPG performance, the presence of the Chevrolet Cruze eco models and the Honda Civic HF sedan showcases that non-diesel vehicles can still achieve notable fuel efficiency and secure positions among the top 30 most fuel-efficient cars. Additionally, the exclusion of hybrid and turbocharged vehicles, as well as the absence of CNG and E85 cars in the top 30 list, highlights the specific focus on conventional fuel-powered cars in this analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://corgis-edu.github.io/corgis/datasets/csv/cars/cars.csv"
df = pd.read_csv(url)

# Filter out rows where 'Engine Information.Engine Type' contains the word 'hybrid'
filtered_data = df[~df['Engine Information.Engine Type'].str.contains('hybrid', case=False)]

# Sort the filtered data by 'Fuel Information.Highway mpg' in descending order
sorted_data = filtered_data.sort_values(by='Fuel Information.Highway mpg', ascending=False)

# Select the top 30 cars based on 'Fuel Information.Highway mpg'
top_thirty_cars = sorted_data.head(30)

# Create the bar chart with different colors based on 'Fuel Information.Fuel Type' and 'Engine Information.Engine Type'
colors = []
for fuel_type, engine_type in zip(top_thirty_cars['Fuel Information.Fuel Type'], top_thirty_cars['Engine Information.Engine Type']):
    if 'diesel' in engine_type.lower() or fuel_type == 'Diesel':
        colors.append('lightgrey')
    elif fuel_type == 'Compressed Natural Gas':
        colors.append('green')
    elif 'turbo' in engine_type.lower():
        colors.append('orange')
    else:
        colors.append('blue')

plt.figure(figsize=(14, 6))
plt.bar(top_thirty_cars['Identification.ID'], top_thirty_cars['Fuel Information.Highway mpg'], color=colors)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Car ID')
plt.ylabel('Highway mpg')
plt.title('Top 30 Cars with Highest Highway mpg (Color-Coded by Fuel and Engine Type)')
plt.tight_layout()
plt.show()
```

Through the examination of the top vehicles for miles per gallon (MPG), it became evident that two standout cars, the 2011 Chevrolet Cruze eco and the 2012 Chevrolet Cruze eco, are equipped with turbochargers. These vehicles, despite being gasoline-powered and non-hybrid, hold prominent positions among a list primarily dominated by diesel cars. This uniqueness in the top ranks can be attributed to the presence of turbochargers, which enhance the engine's efficiency and performance. Another notable outlier in the top MPG list is the 2012 Honda Civic HF sedan, which lacks both a turbocharger and a hybrid system.

The data further reveals that the 2012 Chevrolet Sonic hatchback 1LT turbo and the 2012 Chevrolet Sonic hatchback 1 LTZ turbo also rank near the top of the MPG chart and are the only other two cars in the top 30 with turbocharged engines.

Turbochargers serve as an intriguing topic due to their potential impact on increasing engine efficiency and MPG. They allow for higher horsepower and torque output from smaller engines, which can positively influence fuel economy. However, the data and the graph illustrate that while turbochargers can enhance engine performance, their presence does not necessarily result in a significant boost in MPG. Many of the top-performing gasoline cars in the MPG chart do not feature turbochargers, indicating that achieving fuel efficiency is not solely reliant on turbocharging. Although turbochargers offer advantages in terms of power, they are not extensively used in the efficiency market.

In conclusion, turbochargers play a role in enhancing engine power and performance, and they have contributed to the presence of gasoline cars in the top MPG rankings. However, achieving higher fuel efficiency is a multifaceted process, and turbocharging alone does not guarantee a significant increase in MPG. Despite this, it is promising to see that turbochargers continue to be utilized in certain car models, maintaining a balance between performance and efficiency.

# Worst Performers Chart:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://corgis-edu.github.io/corgis/datasets/csv/cars/cars.csv"
df = pd.read_csv(url)

# Sort the data by 'Fuel Information.Highway mpg' in ascending order
sorted_cars = df.sort_values(by='Fuel Information.Highway mpg', ascending=True)

# Select the worst 30 performers based on 'Fuel Information.Highway mpg'
worst_thirty_cars = sorted_cars.head(30)

# Create a list of colors for each car based on 'Fuel Information.Fuel Type'
colors = ['green' if fuel_type == 'E85' else 'red' for fuel_type in worst_thirty_cars['Fuel Information.Fuel Type']]

# Create the bar chart with conditional coloring
plt.figure(figsize=(12, 6))
plt.bar(worst_thirty_cars['Identification.ID'], worst_thirty_cars['Fuel Information.Highway mpg'], color=colors)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Car ID')
plt.ylabel('Highway mpg')
plt.title('Worst 30 Performers - Highway mpg Comparison')
plt.tight_layout()
plt.show()
```
